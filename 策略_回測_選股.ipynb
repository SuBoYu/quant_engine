# 更新持股投資張數

```python
from finlab.backtest import portfolio

data = Data()

# 利用 mystrategy2 來產生股票清單 stocks
stocks = strategy(data)

# 用portfolio來幫忙計算，給定 1000000 元，依照今天收盤價，股票張數要如何分配
p, total_invest_money = portfolio(stocks.index, 1000000, data) #還可以設定最低手續費lowest_fee Default是20 , 卷商折扣 discount Default是0.6, 增加股票多樣性的成本 addcost Default是10

# 印出股票資訊
print('---------------')
print('|  portfolio  |')
print('---------------')
print(p)
print('total cost')
print(total_invest_money)

#最好還是要有7檔以上的股票
#因為屬於長線的策略，所以用平價買入即可
```
